```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#its a list but we wont put the in " " because we want the value of those.
game_images =[rock,paper,scissors]
player1_name = input("Player1 name: ")
player2_name = input("Player2 name: ")
player1_choice = int(input(f"{player1_name} what do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))

#to hide what player1 chose
print("\n" *50)


player2_choice = int(input(f"{player2_name} what do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))

#choices of players
print(f"{player1_name} chose")
print(game_images[player1_choice])
print(f"{player2_name} chose")
print(game_images[player2_choice])



#will tell about the invalid number
if player1_choice >=3 or player1_choice < 0:
  print("You typed an invalid number.")

#different case as scissors cant beat rock
elif(player1_choice == 0 and player2_choice == 2):
  print(f"{player1_name} wins!")
#different case
elif(player2_choice ==0 and player1_choice == 2):
  print(f"{player2_name} wins!")

#for normal case as scissor beat ppr and ppr beats rock
elif(player1_choice > player2_choice):
  print(f"{player1_name} wins!")

elif(player2_choice > player1_choice):
  print(f"{player2_name} wins!")

elif(player1_choice == player2_choice):
  print(f"It's a draw!")
```
